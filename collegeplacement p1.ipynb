```python
import pandas as pd

df = pd.read_csv("CollegePlacement.csv")

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df = df.drop_duplicates()

print("Shape after cleaning:", df.shape)
print("Missing values after cleaning:\n", df.isnull().sum())
print("Number of duplicate rows after cleaning:", df.duplicated().sum())
print(df)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("CollegePlacement.csv")

df = df.fillna(df.mean(numeric_only=True))
for col in df.select_dtypes("object"):
    df[col] = df[col].fillna(df[col].mode()[0])
df = df.drop_duplicates()


num_cols = df.select_dtypes(np.number)


df_norm = (num_cols - num_cols.min()) / (num_cols.max() - num_cols.min())


df_std = (num_cols - num_cols.mean()) / num_cols.std()


df_encoded = pd.get_dummies(df, drop_first=True)


print("Summary Statistics:\n", df.describe())
print("\nMeans:\n", num_cols.mean())
print("\nMedians:\n", num_cols.median())
print("\nStandard Deviations:\n", num_cols.std())
print("\nCorrelation Matrix:\n", num_cols.corr())


```
